```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Create an empty graph
G = nx.Graph()
```

```python
#Declare nodes
G.add_nodes_from(["Jane", "Ike", "Heather", "Fernando", "Carol", "Andre", "Beverly", "Ed", "Garth", "Diane"])
```

```python
#Declare edges
G.add_edges_from([("Jane", "Ike"), 
                  ("Ike", "Heather"),
                 ("Heather", "Fernando"),
                 ("Fernando", "Carol"),
                  ("Carol", "Andre"),
                  ("Andre", "Beverly"),
                  ("Beverly", "Ed"),
                  ("Ed", "Garth"),
                  ("Heather", "Garth"),
                  ("Garth", "Fernando"),
                  ("Fernando", "Andre"),
                  ("Garth", "Beverly")
                 ])

G.add_edges_from([("Fernando", "Diane"),
                  ("Carol", "Diane"),
                  ("Andre", "Diane"),
                  ("Beverly", "Diane"),
                  ("Ed", "Diane"),
                  ("Diane", "Garth")
                 ])
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
#Create kite from nx function
kkg1 = nx.krackhardt_kite_graph(create_using=None)
nx.draw(kkg1, with_labels=True, font_weight='bold')
```

```python
#add name labels
names = {
    0: "Andre",
    1: "Beverley",
    2: "Carol",
    3: "Diane",
    4: "Ed",
    5: "Fernando",
    6: "Garth",
    7: "Heather",
    8: "Ike",
    9: "Jane",
}

kkg2 = nx.krackhardt_kite_graph(create_using=None)
kkg2 = nx.relabel_nodes(kkg2, names)

nx.draw(kkg2, with_labels=True, font_weight='bold')
```
